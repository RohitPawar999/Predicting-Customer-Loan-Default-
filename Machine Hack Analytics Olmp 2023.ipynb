```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel



# Initialize the Random Forest classifier
clf = RandomForestClassifier()

# Fit the classifier to data
clf.fit(X_train, y_train)

# Use feature importances for feature selection
sfm = SelectFromModel(clf, threshold=0.1)  
sfm.fit(X_train, y_train)

# Transform  data to keep only the selected features
X_train_selected = sfm.transform(X_train)
```

```python
X_test_selected = sfm.transform(X_test)
```

```python
# Initialize a new classifier (or use the existing one)
clf_selected = RandomForestClassifier(n_estimators=130,max_features='sqrt')

# Fit the classifier to the selected feature training data
clf_selected.fit(X_train_selected, y_train)
```

```python
# Make predictions on the test data with selected features
y_pred_selected = clf_selected.predict(X_test_selected)

# Calculate accuracy for the model with selected features
accuracy_selected = accuracy_score(y_test, y_pred_selected)
print("Accuracy with selected features:", accuracy_selected)
```

```python

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
# Train separate AdaBoost models for each label
adaboost_models = []
for label in y_train.columns:
    adaboost_model = AdaBoostClassifier(n_estimators=100, random_state=42)
    adaboost_model.fit(X_train, y_train[label])
    adaboost_models.append(adaboost_model)

# Make predictions for each label
y_pred = pd.DataFrame(columns=y_test.columns)
for i, adaboost_model in enumerate(adaboost_models):
    y_pred[y_test.columns[i]] = adaboost_model.predict(X_test)

# Calculate accuracy for each label
accuracies = []
for label in y_test.columns:
    accuracy = accuracy_score(y_test[label], y_pred[label])
    accuracies.append(accuracy)
    print(f"Accuracy for {label}: {accuracy:.4f}")

# Calculate average accuracy
average_accuracy = sum(accuracies) / len(accuracies)
print(f"Average Accuracy: {average_accuracy:.4f}")
```

```python

# Make predictions using the trained models
new_predictions = pd.DataFrame(columns=['1', '2'])  # Create a DataFrame to store predictions
for i, model in enumerate(adaboost_models):
    new_predictions[new_predictions.columns[i]] = model.predict(data)
```

```python
# Specify the output CSV file path
output_file = 'output_data_7.csv'

# Save DataFrame to a CSV file without including the header
new_predictions.to_csv(output_file, index=False, header=False, mode='a')  # mode='a' to append to the file

# Manually add the specific column names as the first row
custom_column_names = "'primary_close_flag', 'final_close_flag'"
with open(output_file, 'r+') as f:
    content = f.read()
    f.seek(0, 0)
    f.write(custom_column_names + '\n' + content)

print(f"DataFrame saved to {output_file} with manually added column names")
```
